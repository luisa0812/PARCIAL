##  Diferencia entre regresion y modelo 

<p style='text-align: justify;'> Un modelo es una representación simplificada, a través de ecuaciones, funciones o fórmulas matemáticas, de un fenómeno o de la relación entre dos o más variables.
<p style='text-align: justify;'> Regresion busca determinar la relación entre una variable dependiente (Y) con respecto a otras variables llamadas explicativas o independientes (X).  con el fin de determinar si existe o no relación causal entre una variable dependiente (Y) y un conjunto de otras variables explicativas (X). Asimismo, el modelo busca determinar cuál será el impacto sobre la variable Y ante un cambio en las variables explicativas (X).

## REGRESIÓN DE FALLECIDOS 


```python
muertes.values.shape # a nuestra lista de datos le da dimensiones en la matriz
```

```python
y=muertes.values #lista de datos de los fallecidos
y.shape #  dimensión en la matriz en y
```

```python
x=np.arange(len(y)).reshape(166,1)# reshape nos permite dar una nueva dimensión a la matriz sin que nos cambie los datos
print(x.shape) #  dimensión en la matriz en y
```

```python
import numpy as np # utilizamos la libreria numpy y lo nombramos np
y=muertes.values.reshape(len(muertes))# reshape nos permite dar una nueva dimensión a la matriz sin que nos cambie los datos
print(y.shape) #  dimensión en la matriz en y
x=np.arange(len(y))#nos permite correr los datos de y
print (x.shape) #  dimensión en la matriz en x

from scipy.optimize import curve_fit # scipy.optimize nos proporciona que las funciones se puedan minimizar (o maximizar) en funciones objetivo, curve_fit nos permite ver los datos en cuadratica 

## modelo para plantear la regresion 

def muertos(x,a,b,c,d): #Ecuación 
    return a*x**3+b*x**2+c*+d # nos permite que corra la ecuación


## Regresiòn
pars,regresiòn=curve_fit(muertos,x,y) # parametros que utilizamos en la ecuacion anterior que es cuadratica
print(pars)
```

```python
import pylab as pl # con esta libreria se puede establecer limites en las graficas
plt.plot(x,y,"o")
plt.plot(x,muertos(x,*pars))
plt.title("Regresión de fallecidos  ") # nos permite darle un titulo a la regresión
pl.xlim(x.min() * 0.0, x.max() * 0.99)# con xlim se esta-.blece el  limite en el eje x
plt.xlabel('x')# nos permite darle el nombre al eje x
plt.ylabel('y')# nos permite darle el nombre al eje y

plt.show()
```

```python
# Matrix A
import numpy as np
B=np.random.randint(low=1,high=6,size=(300,300))# Esta función podemos hacer numeros aleatorios
print("B =","\n",B,"\n")

#Vector
c=np.random.randint(low=6,high=9,size=(300,1))# Esta función podemos hacer numeros aleatorios
print("c =","\n",c,"\n")

# By=c

y= B@c
print("y =","\n",y,"\n")
print(y[y>0.5])
```
